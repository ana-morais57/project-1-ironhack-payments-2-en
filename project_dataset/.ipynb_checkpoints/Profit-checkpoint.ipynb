The main objective is to analyze **user cohorts defined by the month of creation of their first cash advance**. You will **track the monthly evolution of key metrics** for these cohorts, enabling IronHack Payments to gain valuable insights into user behavior and the performance of their financial services.

# Metrics to analyze

### You will calculate and analyze the following metrics for each cohort:

- Frequency of Service Usage:
    - Understand how often users from each cohort utilize IronHack Payments' cash advance services over time.

    - For this, we only need to look at the cash_requests .csv file, and see the frequency of cash requests over time

    We are assuming that the creation of the first cash advance is the date of the Credit Request creation

- Incident Rate:
    - Determine the incident rate, specifically focusing on payment incidents, for each cohort. Identify if there are variations in incident rates among different cohorts.

- Revenue Generated by the Cohort:
    - Calculate the total revenue generated by each cohort over months to assess the financial impact of user behavior.

```python
# Let's assign users to cohorts based on the month of their first accepted cash request
cash_df['cohort'] = cash_df.groupby('user_id')['created_at'].transform(lambda x: x.min().strftime('%Y-%m'))
cash_df.cohort
```

```python
# Verify cohort assignment
print(cash_df[['user_id', 'created_at', 'cohort']].head())
```

```python
# Extract month of cash request for analysis
cash_df['request_month'] = cash_df['created_at'].dt.to_period('M')
# Count cash requests by cohort and month
cohort_usage = cash_df.groupby(['cohort', 'request_month']).size().unstack(fill_value=0)

# Plot frequency heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cohort_usage, annot=True, fmt="d", cmap="Blues")
plt.title("Frequency of Service Usage by Cohort")
plt.xlabel("Request Month")
plt.ylabel("Cohort")
plt.show()
```

```python
cash_df.nunique()
```

```python
# Link fees to cash requests
merged_df = pd.merge(fees_df, cash_df, left_on='cash_request_id', right_on='id', how='inner')

# Filter for incidents
incident_fees = merged_df[merged_df['type'] == 'incident']

# Identify users with incidents
users_with_incidents = incident_fees['user_id'].unique()

# Calculate incident rate per cohort
incident_rate = cash_df.groupby('cohort')['user_id'].apply(
    lambda x: np.mean(x.isin(users_with_incidents))
)

# Visualize incident rate
incident_rate.plot(kind='bar', figsize=(10, 6), color='coral')
plt.title("Incident Rate by Cohort")
plt.xlabel("Cohort")
plt.ylabel("Incident Rate")
plt.show()
```

```python
# Filter fees for accepted status
accepted_fees = fees_df[fees_df['status'] == 'accepted']

# Link fees to cohorts via cash requests
accepted_fees = pd.merge(accepted_fees, cash_df[['id', 'cohort']], 
                         left_on='cash_request_id', right_on='id', how='inner')

# Sum revenue by cohort and month
accepted_fees['paid_month'] = accepted_fees['paid_at'].dt.to_period('M')
cohort_revenue = accepted_fees.groupby(['cohort', 'paid_month'])['total_amount'].sum().unstack(fill_value=0)

# Plot revenue heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cohort_revenue, annot=True, fmt=".2f", cmap="Greens")
plt.title("Revenue by Cohort")
plt.xlabel("Paid Month")
plt.ylabel("Cohort")
plt.show()
```
